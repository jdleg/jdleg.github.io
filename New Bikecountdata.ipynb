the imbalance between them (arrivals - departures), and total usage and imbalance for each station over 8 months of data.

<b>What the functions below do</b>

The <b>fullstation</b> function below automates what a couple of functions were doing.<br>
It takes a <b>station id</b> and a <b>bike data frame</b> as arguments, and calls <b>stationframes</b> and <b>countsmatrixcreate</b>.<br>
<b>stationframes</b> subselects trips that involve arriving at or departing from the specified station and calls <b>hourcount</b><br>
<b>hourcount</b> generates average numbers of arrivals and departures for all 168 hours of the week.<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;<i>(In other words, if the given data spans 4 Mondays, it will sum all the trips on Mondays from 3-4 pm, divide them by 4, and return that value for 3-4 pm on Mondays.)</i>
    
<b>fullstationlist</b> is almost the same function, but it takes in a list of station id's and calls stationframeslist, which collect records for all those id's, and generates hourly averages for a set of stations.

A related idea would be to use geographical information to select stations that are close by, and thus experience similar patterns of demand.

I've made versions of these files to work 16 weekday/weekend x 3-hour period values instead of 168. For instance, it will sum all the trips made from 3-6 pm Monday through Friday, and return the average value across all such 3-hour periods as the result for weekdays 3-6 pm.
